## Structure of Scraped Object

```python
trial = scraper.get_tweets('ONEPIECE', mode='hashtag', number=20)
```

```python
print(f'Type of Scraped object: {type(trial)}\n')
print(f'Keys of Scraped object: {trial.keys()}\n')
print(f"Type of object 'tweets': {type(trial['tweets'])}\n")
print(f"Type of object a 'tweet': {type(trial['tweets'][0])}\n")
print("Keys  and Value Types of a 'tweet':")
t = trial['tweets'][0]
for key, val in t.items():
    print(f"Key: {key}, Type: {type(val)}")
print('\n')
```
